```python
from multi_task_nlp.token_classification import SlotFillingDataModule
import pytorch_lightning as pl
from transformers.data.data_collator import pad_without_fast_tokenizer_warning
```

```python
# Set random seed for reproducibility
pl.seed_everything(42)

# Login to W&B
# wandb.login()

# Parameters
MODEL_NAME = "distilbert/distilbert-base-uncased"
BATCH_SIZE = 32
MAX_EPOCHS = 1
MAX_STEPS = -1  # 30
LEARNING_RATE = 2e-5

# Initialize data module
data_module = SlotFillingDataModule(model_name=MODEL_NAME, batch_size=BATCH_SIZE, max_length=128, val_split=0.15)
data_module.setup()
```

```python
batch = {k: v for k, v in data_module.train_dataset[:].items() if k != "label_ids"}
batch = pad_without_fast_tokenizer_warning(
    data_module.tokenizer,
    batch,
    padding=True,
    # max_length=data_module.max_length,
    return_tensors="pt",
)
batch
```
